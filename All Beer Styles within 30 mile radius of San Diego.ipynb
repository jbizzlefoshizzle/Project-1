```python
from brewerydb import *
from config import api_key
import requests
import pandas as pd
```

```python
BreweryDb.configure(api_key, DEFAULT_BASE_URI)
```

```python
sd_styles = BreweryDb.styles({'geo':1,
                         'lat':32.7157,
                         'lng':117.1611,
                         'radius':30,
                         'units':'m'})
# What is the difference between
# BEER STYLE, and
# BEER CATEGORY?
```

```python
beer_styles =[]
for y in range(len(sd_styles["data"])):
    print(f" printing beer styles index # {y} within 30 mile radius of SD")


    beer_styles.append(sd_styles['data'][y]['name'])
```
