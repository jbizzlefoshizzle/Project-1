```python
from brewerydb import *
from config import api_key
import requests
import pandas as pd
```

```python
BreweryDb.configure(api_key, DEFAULT_BASE_URI)
```

```python
sd_ingredients = BreweryDb.ingredients({'geo':1,
                         'lat':32.7157,
                         'lng':117.1611,
                         'radius':30,
                         'units':'m'})
```

```python
#Looping through beer names an categories
beer_name =[]
beer_category = []
for y in range(len(sd_ingredients["data"])):
    print(f" printing beer names and categ index # {y} within 30 mile radius of SD")


    beer_name.append(sd_ingredients['data'][y]['name'])
    beer_category.append(sd_ingredients['data'][y]['category'])
```

```python
# conveting lists to dictionary
sd_ingredients_dcit = {"Beer Name":beer_name,
                       "Beer Category":beer_category
                      }
```

```python
#converting dictionary to DataFrame
sd_ingredients_df = pd.DataFrame.from_dict(sd_ingredients_dcit)
sd_ingredients_df
```
